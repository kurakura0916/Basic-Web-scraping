```python
import re
import time
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
rating_score_list =[]
review_list =[]

# 1~4ページまでのレビュー内容とスコア（5点満点）をスクレイピング
for i in range(1,4):
    domain = "https://filmarks.com/movies/65813" # 対象の映画のページ
    url = domain + "?page={i}".format(i=1) # ページのパラメータを含んだURL
    r = requests.get(url)
    soup = BeautifulSoup(r.content,"lxml")

    content = soup.find("div",class_="p-main-area p-timeline")
    persons = content.find_all("div",class_="js-mark-card p-mark")

    for person in persons:
        rating_score_list.append(person.find("div",class_="c-rating__score").text)
        review_list.append(person.find("div",class_="p-mark__review").text)

    time.sleep(1)
    
df = pd.DataFrame({'review':review_list,
                                    'rating_score':rating_score_list}, 
                                    columns=['review','rating_score'])
```
